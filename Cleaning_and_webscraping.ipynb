### Web scraping:

```python
from bs4 import BeautifulSoup
import requests
```

```python
# 2. find url and store it in a variable
url = "https://en.wikipedia.org/wiki/List_of_countries_by_meat_consumption#cite_note-FAO2017-12"
```

```python
# 3. download html with a get request
response = requests.get(url)
```

```python
response.status_code # OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
# Find the table in html
table = soup.find('table', class_='wikitable sortable')


if table:
    # Extractin data from table
    data = []
    for row in table.find_all('tr'):
        row_data = []
        for cell in row.find_all(['th', 'td']):
            row_data.append(cell.text.strip())
        data.append(row_data)

    # Imprimir los datos extraídos
    for row in data:
        print(row)
else:
    print("Impossible to find the table.")
```

```python
# Now we create lists for the columns
Country_Dependency = []
kg_person_2002 = []
kg_person_2009 = []
kg_person_2017 = []
kg_person_2020 = []


if table:
    # In the table from second line
    for row in table.find_all('tr')[1:]:
        # to get all the cells
        cells = row.find_all(['th', 'td'])
        
        if cells:
            # Compleate the lists with data
            Country_Dependency.append(cells[0].text.strip())
            kg_person_2002.append(cells[1].text.strip())
            kg_person_2009.append(cells[2].text.strip())
            kg_person_2017.append(cells[3].text.strip())
            kg_person_2020.append(cells[4].text.strip())

# New DataFrame
wikitable = pd.DataFrame({
    "Country/Dependency": Country_Dependency,
    "kg/person (2002)[9][note 1]": kg_person_2002,
    "kg/person (2009)[10]": kg_person_2009,
    "kg/person (2017) [11]": kg_person_2017,
    "kg/person (2020)[12]": kg_person_2020
})


print(wikitable)
```
